```python
# Import required libraries
import pandas as pd         
import numpy as np          
from plotnine import *
```

```python
# Load the dataset
coffee = pd.read_csv("coffe.csv")
```

```python
# Visualization 1: Sales count by coffee type
viz1 = (
    ggplot(coffee, aes(x="coffee_name")) 
    + geom_bar(fill="steelblue") 
    + labs(
        title="Number of Sales by Coffee Type",
        x="Coffee Type",
        y="Count of Sales"
    )
    + theme_minimal()
)
viz1.show()
```


This plot shows how many times each type of coffee was sold and Americano with Milk was the highest purchased followed by Latte. It allows us to see which drinks are most popular with customers.


```python
# Visualization 2: Average money spent by Time of Day

viz2 = (
    ggplot(coffee, aes(x="Time_of_Day", y="money", fill="Time_of_Day")) 
    + stat_summary(fun_y=np.mean, geom="bar") 
    + labs(
        title="Average Spending by Time of Day",
        x="Time of Day",
        y="Average Money Spent"
    )
    + theme_minimal()
)
viz2.show()
```

```python

This plot shows the average amount spent by customers depending on whether they visited in the Morning, Afternoon, or Evening. The highest coffee purchased was during night.

```
